```python

df = pd.read_csv("./SSI ACA Data/PANEL_SSI.csv")
df = df[df['State']!="United States"]
df = df.drop("Unnamed: 0", axis=1)
```

```python
NEregion = df[df['RegName']=="Northeast Region"].pivot_table(values='d_pct_BenePerCap', index = "Year", columns="State")
MWregion = df[df['RegName']=="Midwest Region"].pivot_table(values='d_pct_BenePerCap', index = "Year", columns="State")
Sregion = df[logical_and(df['RegName']=="South Region", df["State"]!="District of Columbia")].pivot_table(values='d_pct_BenePerCap', index = "Year", columns="State")
Wregion = df[df['RegName']=="West Region"].pivot_table(values='d_pct_BenePerCap', index = "Year", columns="State")
```

```python
neplot = NEregion.plot(title="% Change in Disability Recipients, Northeastern Region")
neplot.axvspan(2013, 2015, color='b', alpha=0.1, lw=0)
neplot.legend(ncol=2, fontsize='x-small', loc='upper right')
```
