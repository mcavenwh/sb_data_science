```python
%pylab inline 

import numpy as np
import pandas as pd
import scipy.stats as stats
#import statsmodels.api as sm
#from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df = pd.read_csv("./SSI ACA Data/NATL_SSI.csv")
plt.plot(df.Year, df.Beneficiaries/1000000)
plt.xlabel("Year")
plt.xticks(np.arange(1996, 2015, step=2))
plt.yticks(np.arange(8,14, step = 1))
plt.ylabel("Number of Beneficiaries (millions)")
plt.title("SSI Beneficiaries, Nationwide, 1998-2015")
plt.show()
```
